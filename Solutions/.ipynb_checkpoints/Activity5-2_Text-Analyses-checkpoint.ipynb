## 3. Preprocess the data.
- Use the code chunk from the lecture where we run it all together.
- Delete the section where the data is loaded in again
    - *Hint: You can change all the `cnn` instances to `cnn` by pressing `control (command) F`, typing in `cnn`, hit the arrow to the left of your `b`, type `cnn` into the replace box, hit `Replace All`.*
- Notice how our `preprocessor` (from `pip install tweet-preprocessor`) handles our RT's and @'s.

```python
import preprocessor as p
from gensim.parsing.preprocessing import remove_stopwords

# make a function that does the stopword function and the preprocessing cleaning (remove hyperlinks)
def preprocess_tweet(row):
    text = row['tweet']
    text = remove_stopwords(text) # remove stopwords
    text = p.clean(text) # remove links 
    return text

# run the function
cnn['tweet'] = cnn.apply(preprocess_tweet, axis=1) # you have to run the function first!!!

#do the other bits
cnn["tweet"] = cnn.tweet.str.lower() # make all lower case
cnn["tweet"] = cnn.tweet.str.replace('[^\w\s]',' ').str.replace('\s\s+', ' ') # whitespace and numbers
cnn
```

## 4. Analyze the data.
- Make the two small functions from `textblob` that run the `sentiment.subjectivity` and `sentiment.polarity` attributes.
- Use the apply function to make two new columns.
- Display your updated dataframe.

```python
# pip install textblob
from textblob import TextBlob

# Create a function to get the subjectivity
def getSubjectivity(text):
   return TextBlob(text).sentiment.subjectivity

# Create a function to get the polarity
def getPolarity(text):
   return  TextBlob(text).sentiment.polarity


# Create two new columns 'Subjectivity' & 'Polarity'
cnn['Subjectivity'] = cnn['tweet'].apply(getSubjectivity)
cnn['Polarity'] = cnn['tweet'].apply(getPolarity)

# Show the new dataframe with columns 'Subjectivity' & 'Polarity'
cnn
```

## 5. Analyze the Polarity data.

- Use the function `getAnalysis` from the lecture to make the `Polarity` variable meaningful.
- Update your dataframe and add a new column `Analysis`. 
- Display your updated dataframe.

```python
# Create a function to compute negative (-1), neutral (0) and positive (+1) analysis
def getAnalysis(score):
    if score < 0:
      return 'Negative'
    elif score == 0:
      return 'Neutral'
    else:
      return 'Positive'
cnn['Analysis'] = cnn['Polarity'].apply(getAnalysis)
# Show the dataframe
cnn.where(cnn.Analysis == "Negative").dropna()
```

## 6. Make a wordcloud
- Visualize your dataframe.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

allWords = ' '.join([twts for twts in cnn['tweet']])
wordCloud = WordCloud(width=500, height=300, random_state=21, max_font_size=110).generate(allWords)


plt.imshow(wordCloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```
