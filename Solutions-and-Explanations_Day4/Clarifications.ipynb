```python
import pandas as pd
sleep = pd.read_csv('sleep.csv')
list(sleep.columns)
```

```python
import statsmodels.api as smf
lm = smf.formula.ols("price ~  city", data = sleep).fit()
print(lm.summary())
```

```python
import statsmodels.api as smf
lm = smf.formula.ols("price ~  city", data = sleep).fit()
print(lm.summary())
```

# More on `pivot_table()`

- Run the function `.columns` on your two tables.
    - Notice how 'Happiness score' is the only columns displayed. This is because 'Region' and 'Year' are used as indices.
