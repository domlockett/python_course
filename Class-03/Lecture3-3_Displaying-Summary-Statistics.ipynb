## The bar chart

```python
# most recent first!
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

happiness = pd.read_csv('happiness.csv')
happiness
```

```python
pd.pivot_table(happiness, index= 'Region',columns= "Year", values= "Happiness score").plot(kind= 'bar')
plt.ylabel("Happiness Rank")
```

```python
# that graph is not the most visually appealing

import seaborn as sns # has nicer default themes

sns.set()
pd.pivot_table(happiness, index= 'Region', columns= 'Year', values= "Happiness score").plot(kind= 'bar')
plt.ylabel("Happiness Rank")
```

```python
our_data = pd.pivot_table(happiness, index= 'Region', columns= 'Year', values= "Happiness score")
our_data
```

```python
plt.figure()
```

```python
# better but not the best!
# this all has to be run in the same code chunk so that it knows all of the data being analyzed

f= plt.figure() # our plot is regarded as a figure so when we want to make minute adjustments we want to go ahead and make an object using plt.figure()
plt.title("Happiness Rank") # here we set the title of the figure we will plot
our_data.plot(kind="bar", ax= f.gca()).legend(bbox_to_anchor=(1,1)) #.gca means get current axes and it directs us towards the current figure and its axes

#legend() takes multiple arguments. bbox_to_anchor tells on which axes  to display our legend
```

```python
f= plt.figure() 

plt.title("Happiness Rank") 

our_data.plot(kind="bar", ax= f.gca()).legend(bbox_to_anchor=(0,1)) #notice how changing bbox changes legend location
```

```python
f= plt.figure() 
plt.ylabel("Happiness Rank") 
plt.title("Happiness by year and region")
our_data.plot(kind="bar", ax= f.gca()).legend(bbox_to_anchor=(1,1))
plt.xticks(rotation=45, ha='right') # This has to be called after plotting!
```

## The line chart
- Fertility data was retrieved from The World Bank.
- Here we go back to our pivot method and look at a nice line.

```python
import numpy as np
fertility = pd.read_csv("fertility.csv")
countries = np.load('data\countries.npy')
fertility2015 = fertility[fertility.Year == 2015]
fertility = fertility.dropna()
fertility = fertility[fertility["Country Name"] != "World"]
fertility = fertility[fertility["fertility"] != ".."]
fertility = fertility[fertility["Country Name"].isin(countries)]

fertility.fertility = fertility.fertility.astype(float)
fertility
```

```python
fertility_avg = pd.pivot_table(fertility, index= 'Year',  values= "fertility")
fertility_avg
```

```python

fertility_avg.plot(kind= 'line')
plt.title("Average Fertility by Year")
plt.ylabel("Fertility Rate")
```

## The scatter plot


```python
import seaborn as sns # has nicer default themes

sns.set()

fertility.plot.scatter(x="fertility", y="Year", alpha =.2, color= "DarkRed")
plt.title("Fertility Rate by Year") 
```

### Let's combine the two datasets to look at some data
-- Infant mortality data was retrieved from [The World Bank](https://data.worldbank.org/indicator/SP.DYN.TFRT.IN).

```python
infant = pd.read_csv("data\infant_mortality.csv") # per 1000 live births
infant = infant.rename(columns = {"2015 [YR2015]": "inf_mortality"})
infant = infant[["Country Code","inf_mortality"]]

```

```python
infant
```

```python
inf_fert = pd.merge(infant, fertility2015, how = "outer")
inf_fert
```

```python
#plot an obvious relationship
inf_fert.plot.scatter(x="fertility", y="inf_mortality" , color = "DarkRed")
plt.title("Infant Mortality by Fertility Rate") 
```

```python
inf_fert.fertility = inf_fert.fertility.replace('..',np.nan)
inf_fert.fertility = inf_fert.fertility.astype(float)

inf_fert.plot.scatter(x="fertility", y="inf_mortality" , color = "DarkRed")

plt.title("Infant Mortality by Fertility Rate") 
```

## Exploring data

- This data was scraped by Paul Boal from the [Clear Health Costs](https://clearhealthcosts.com/) website.

    - It is a collection of prices for dental related procedures from 4 cities: New York, San Francisco, Miami, and Dallas.

### Load the data

```python
dental =pd.read_csv('data\dental_data.csv')
dental = dental.drop('Unnamed: 0', axis = 1)
dental
```

## Inspect the data

- We can look at the distribution of prices.

    - If we are not comfortable eyeballing the type of distribution and need this for later analyses, we can use tools in the `SciPy` package.

```python
import seaborn as sns
import scipy # math science and engineering module
sns.distplot(dental.price)
```

```python
# Doesnt look normal but lets check
sns.distplot(dental.price, fit=scipy.stats.norm)
```

```python
#Check gamma
sns.distplot(dental.price, fit=scipy.stats.cauchy)
```

```python
# logistic?
sns.distplot(dental.price, fit=scipy.stats.gamma)
```

## Remove bad data

- The data shows we have an odd number of zero entries so lets dump those.

```python
total = len(dental)
zeros = sum(dental.price == 0)
print("{} out of {} have a  price of $0".
      format(zeros, total))
```

```python
dental = dental[dental.price != 0]
dental.head()
```

```python
sns.distplot(dental.price, fit=scipy.stats.gamma)
```

## Looking at the trends of data

- There's a large skew at the low end of prices, so let's breakdown the data by which zip code we searched around.  That will show us if the bubble is specific to one region.

- *Note: The violin plot below is an estimate of the probability density function (PDF) which relies on the non-parametric guassian kernel density estimator (KDE) which is why we see values which are outside the actual range. One way to overcome this is to look at a simple box plot or manually set the range for the KDE*

```python
sns.violinplot(data=dental, y='price', x='city')
plt.title("Dental Costs by City") 
```

```python
sns.boxplot(data=dental, y='price', x='city')
```

```python
sns.violinplot(data=dental, y='price', x='city', cut =0)
```
