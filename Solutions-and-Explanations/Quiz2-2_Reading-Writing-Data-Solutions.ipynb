## 2. Manually load in data.
- Manually load in the `preexisting.csv` file.
    - The preexisting.csv is in the file folder data.
    - Use the file call `data\preexisting.csv` to access the data in this other folder.
    - Print the file line by line.

```python
#write your solution to problem 2 below this line:
with open('data\preexisting.csv') as csv:
    for line in csv:    #create a for loop to go through every line in the document. 
        values = line.strip().split(",") # we can use multiple .functions at once
        print(values)
```

## 3. Write in data using the csv module.

    - Import the csv module
    - Load in the entire `preexisting.csv` dataset and print the data object line by line.

```python
#write your solution to problem 3 below this line:

import csv
with open('data\preexisting.csv') as file: # open the file
    for i in csv.reader(file): # use for loop to have the csv function read each line
        print(i) # print to view
        
# you still have to use for loop which is not ideal
```
