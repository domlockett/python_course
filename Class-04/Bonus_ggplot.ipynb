# ggplot in Python



```python
# you will need to use pip install plotnine for this
from plotnine import ggplot, geom_point, aes, stat_smooth, facet_wrap
from plotnine import geom_line, geom_smooth
from plotnine.data import mtcars

(ggplot(mtcars, aes('wt', 'mpg'))
 + geom_point())
```

```python


(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)'))
 + geom_point()
 + stat_smooth(method='lm')
 + facet_wrap('~gear'))
```

```python

# you will have to use pip install vega_datasets

from vega_datasets import data

df = data.iris()

# Create a simple scatter plot.
# Note, the parens wrapping the statement allow you to use `+` at the end of the line
# without escaping with a backslash.
(ggplot(df, aes('petalWidth', 'petalLength')) +
  geom_point())
```

```python
(ggplot(df, aes('petalWidth', 'petalLength')) +
  geom_point(color='darkgreen', size=4)
)
```

```python
(ggplot(df, aes('petalWidth', 'petalLength', color='species')) +
  geom_point() +
  stat_smooth(method='lm')
)
```
